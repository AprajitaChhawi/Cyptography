```python
from sympy.ntheory import randprime
import random
import pandas as pd
def Fiat_Shamir_protocol(p,q,k):
    while(k):
        k=k-1
        n=p*q
        n1.append(n)
        s=random.randint(1,n)
        v=mod(s**2,n)
        v1.append(v)
        r=random.randint(0,n)
        r1.append(r)
        x=mod(r**2,n)
        x1.append(x)
        c=random.randint(0,1)
        c1.append(c) #sends this x to bob as witness
        y=r*(s**c) #alice calculates y and sends it to bob
        res1=mod(y**2,n) #bob calculates res1 and res2 .
        res2=mod(x*(v**c),n) #if both are equal , then alice is honest.
        res11.append(res1)
        res21.append(res2)
        if(res1==res2):
            output1.append("Probable")
        else:
            print("Improbable")
            break
p = randprime(2 ** 10, 2 ** 11)
q = randprime(2 ** 10, 2 ** 11)
column_names = ['x','v','n','r','c','y2 mod n','xv^c mod n','Output']
df = pd.DataFrame(columns = column_names)
x1=[]
v1=[]
n1=[]
r1=[]
c1=[]
res11=[]
res21=[]
output1=[]
Fiat_Shamir_protocol(p,q,10)
df['x']=x1
df['v']=v1
df['n']=n1
df['r']=r1
df['c']=c1
df['y2 mod n']=res11
df['xv^c mod n']=res21
df['Output']=output1
df
```

```python
def euler(p, q):
    return ((p - 1) * (q - 1))
def NOD(a, b): 
    while a != 0 and b != 0:
        if a > b:
            a %= b
        else:
            b %= a
    gcd = a + b  
    return (gcd)
def bezout(a, b):
    oldB = b
    x, xx, y, yy = 1, 0, 0, 1
    while b:
        q = a // b
        a, b = b, a % b
        x, xx = xx, x - xx*q
        y, yy = yy, y - yy*q
    if(x < 0):
        x = x + oldB
    return (x)
def modInverse(a, m): 
    a = a % m 
    for x in range(1, m): 
        if ((a * x) % m == 1): 
            return x 
    return 1
```
